```python
import json
from sympy.parsing.mathematica import parse_mathematica
from sympy import *

def read_json_file(file_path):
    try:
        with open(file_path, 'r') as json_file:
            data = json.load(json_file)
            return data
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
        return []

file_path = '/content/ints_ann.json'  # Replace with your actual file path
json_data = read_json_file(file_path)

print(json_data)

# Extract the prompts from the dictionary
prompts_list = [item["prompt"] for item in json_data["annotations"]]

#print(prompts_list)
```

## **Integral**

```python
import wolframalpha
import requests
import os
import urllib.parse

# App id obtained by the above steps
app_id = 'XTJKXE-UV98JR6EKE'

# Instance of wolf ram alpha
# client class
client = wolframalpha.Client(app_id)

result_list = []


for i in range(len(prompts_list)):
  plaintext = ''
  prompt = urllib.parse.quote(prompts_list[i])
  print(prompt)
  query_url = f"http://api.wolframalpha.com/v2/query?" \
            f"appid={app_id}" \
            f"&input={prompt}" \
            f"&format=plaintext" \
            f"&output=json"

  r = requests.get(query_url).json()

  print(f"Result: {r}")
  if(r["queryresult"]["numpods"]>0):
    for j in range(len(r["queryresult"]["pods"])):
      title = r["queryresult"]["pods"][j]["title"]
      if(title == "Result"):
        data = r["queryresult"]["pods"][j]["subpods"][0]
        plaintext = data["plaintext"]
        print(plaintext)
      elif (title == "Definite integral"):
        data = r["queryresult"]["pods"][j]["subpods"][0]
        text = data["plaintext"].split("=")
        plaintext = text[1].strip()

    print(f"Result: {plaintext}")

  result_list.append(plaintext)

#print(result_list)
```

```python
# Access the integral results
i=0
print(result_list)
for annotation in json_data["annotations"]:
      annotation["integral_result"] = str(result_list[i])
      i=i+1
print(json_data)
# Save the updated JSON data to a file
with open("ints_ann_result.json", "w") as json_file:
    json.dump(json_data, json_file, indent=4)

print("Updated JSON data saved to 'ints_ann_result.json'")
```
